```python
import pandas as pd
df = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
import tweepy

## keys are hidden
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
ids = df.tweet_id.values
len(ids)
```

```python
#writing json data from API to a text file
import json
from timeit import default_timer as timer
count = 0
exceptions = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    for tweet_id in ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            exceptions[tweet_id] = e
end = timer()
print(end - start)
print(exceptions)
```

```python
# load selected json data from text file to a dataframe 
df_list = []
with open('tweet_json.txt') as f:
    for line in f:
        data = json.loads(line)
        tweetid = data['id']
        retweetcount = data['retweet_count']
        favoritecount = data['favorite_count']
        df_list.append({'tweet_id': tweetid,
                        'retweet_count': retweetcount,
                        'favorite_count': favoritecount})
df_extra = pd.DataFrame(df_list, columns = ['tweet_id', 'retweet_count', 'favorite_count'])
```

## Clean

```python
main_clean = df.copy()
breeds_clean = df_breeds.copy()
count_clean = df_extra.copy()
```

```python
# Main: include retweets and replies that are not original tweets
# Define: remove the those non-original records and column names indicating those records
```

```python
# Code
main_clean = main_clean[main_clean['retweeted_status_id'].isnull()]
main_clean = main_clean.drop(['retweeted_status_id', 'retweeted_status_user_id','retweeted_status_timestamp'], axis = 1)
main_clean = main_clean[main_clean['in_reply_to_status_id'].isnull()]
main_clean = main_clean.drop(['in_reply_to_status_id', 'in_reply_to_user_id'], axis = 1)
```

```python
# Test
main_clean.info()
```

```python
# Tidy
# Main: multiple columns for dog stages
# Define: Replace None with null for each dog stage and concatenate the dog stage information into one column. Drop the multiple
# dog stage columns
```

```python
# Code
import numpy as np
for stage in ['doggo','floofer', 'pupper','puppo']:
    main_clean[stage].replace('None', np.nan, inplace=True)
    
main_clean['dog_stage'] = main_clean[['doggo','floofer','pupper','puppo']].fillna('').sum(axis=1)
main_clean.loc[(main_clean.dog_stage == 'doggopupper')| (main_clean.dog_stage == 'doggofloofer')| 
               (main_clean.dog_stage == 'doggopuppo'), 'dog_stage'] = 'multiple'  
main_clean = main_clean.drop(['doggo','floofer','pupper','puppo'], axis = 1)
```

```python
# Test
main_clean.dog_stage.value_counts()
```

```python
# Main: Wrong datatypes
# Define Convert tweet_id to string type, timestamp to datetime and dog_stage to category type
# Cleaned timestamp to the format "YYYY-MM-DD hh-mm-ss"
```

```python
# Code
main_clean.tweet_id = main_clean.tweet_id.apply(str)
breeds_clean.tweet_id = breeds_clean.tweet_id.apply(str)
count_clean.tweet_id = count_clean.tweet_id.apply(str)
main_clean.timestamp = pd.to_datetime(main_clean.timestamp.str[:-6])
main_clean.dog_stage = main_clean.dog_stage.astype('category')
```

```python
#Test:
main_clean.info()
```

```python
# Main: incorrect names that start with a lowercase letter
# Define: convert those incorrect names to 'None'
```

```python
# Code
main_clean.loc[main_clean.name.str.islower(),'name'] = 'None'
```

```python
# Test
main_clean[main_clean.name.str.islower()]
```

```python
# Main: source format
# Define: regex to extract appropriate information for source
```

```python
# Code
main_clean.source = main_clean.source.str.extract('\>(.*)\<',expand = True)
```

```python
# Test
main_clean.source.value_counts()
```

```python
# Main: rating denominators 
# Define: remove records with denominators other than 10
```

```python
# Code
main_clean = main_clean[main_clean.rating_denominator == 10]
```

```python
# Test
main_clean.rating_denominator.value_counts()
```

```python
# Main: rating numerators
# Define: remove records with denominators not in the range of 7 to 14
```

```python
# Code
main_clean[main_clean.text.str.contains(r"(\d+\.\d*\/\d+)")][['text', 'rating_numerator']]
```

```python
main_clean[main_clean.text.str.contains(r"(\d+\.\d*\/\d+)")][['text', 'rating_numerator']].text.str.extract("(\d+\.\d*)\/\d+")
```

```python
main_clean.loc[45, 'rating_numerator'] = 13.5
main_clean.loc[695, 'rating_numerator'] = 9.75
main_clean.loc[763, 'rating_numerator'] = 11.27
main_clean.loc[1712, 'rating_numerator'] = 11.26
```

```python
main_clean = main_clean[(main_clean.rating_numerator >= 7) & (main_clean.rating_numerator <= 14)]
```

```python
# Test
main_clean.rating_numerator.value_counts()
```

```python
# Breeds: Duplicated images and non-dog images
# Define: remove duplicates and exclude non-dog images
```

```python
# Code
breeds_clean = breeds_clean[-breeds_clean.jpg_url.duplicated()]
breeds_clean = breeds_clean[(breeds_clean['p1_dog'] == True) |
                            (breeds_clean['p2_dog'] == True) |
                            (breeds_clean['p3_dog'] == True)]
```

```python
# Test
breeds_clean.info()
```

```python
# Tidy: Drop duplicated columns and unneeded columns
# Define: Merge 3 tables to a master dataframe
```

```python
# Code
temp = pd.merge(main_clean, breeds_clean,
                            on='tweet_id', how='inner')
```

```python
twitter_archive_master = pd.merge(temp, count_clean,
                            on='tweet_id', how='inner')
```

```python
# Test
twitter_archive_master.info()
```

### Store, Analyze & Visualize

```python
twitter_archive_master.to_csv('twitter_archive_master.csv', sep= ',', encoding = 'utf-8')
```

```python
# Top 10 breeds with the most favorite counts and retweet counts
breeds_sum = twitter_archive_master.groupby('p1').sum()
breeds_sum.sort_values('favorite_count', ascending = False).head(10)
```

```python
breeds_sum.sort_values('retweet_count', ascending = False).head(10)
```

As shown above, the two lists are almost the same with golden retriever, labrador retriever and pembroke being the top 3 most popular breeds

```python
import matplotlib.pyplot as plt
fav = breeds_sum.sort_values('favorite_count', ascending = False).head(10)
fav = fav.reset_index()
fav.plot(x='p1', y='favorite_count', kind='bar', figsize=(10,8), legend =None, color = 'g');
plt.ylabel('Favorite Counts')
plt.xlabel('Breeds')
plt.title('Number of Favorite Counts');
```

```python
# How ratings change overall as time goes
records_2015 = twitter_archive_master[twitter_archive_master.timestamp.dt.year == 2015]
records_2016 = twitter_archive_master[twitter_archive_master.timestamp.dt.year == 2016]
records_2017 = twitter_archive_master[twitter_archive_master.timestamp.dt.year == 2017]
```

```python
mean_2015 = records_2015.rating_numerator.mean()
mean_2016 =records_2016.rating_numerator.mean()
mean_2017 =records_2017.rating_numerator.mean()
```

```python
plt.figure(figsize=(10,8))
plt.plot(['2015','2016', '2017'], [mean_2015,mean_2016,mean_2017],color = 'g', marker = 'o', linestyle = 'dashed')
plt.title('Mean Ratings from 2015 - 2017')
plt.xlabel('Year')
plt.ylabel('Average Rating');
```

```python
# Is there a correlation between ratings and favorite counts or retweet counts 
from scipy import stats
ratings = twitter_archive_master.rating_numerator
fav_counts = twitter_archive_master.favorite_count
retweet_counts = twitter_archive_master.retweet_count
slope1, intercept1, r_squared1, p_value1, std_err1 = stats.linregress(ratings, fav_counts)
print('Ratings vs Favorite Counts')
print("slope: %f  intercept: %f  R squared: %f" % (slope1, intercept1, r_squared1))
slope2, intercept2, r_squared2, p_value2, std_err2 = stats.linregress(ratings, retweet_counts)
print('Ratings vs Retweets Counts')
print("slope: %f  intercept: %f  R squared: %f" % (slope2, intercept2, r_squared2))
```

```python
slope2, intercept2, r_squared2, p_value2, std_err2 = stats.linregress(ratings, fav_counts)
```

```python
plt.figure(figsize=(10,8))
plt.plot(ratings, fav_counts, 'o', label='ratings vs favorite counts',color ='g')
plt.plot(ratings, intercept1 + slope1*ratings, 'r', label='fitted line')
plt.legend()
plt.title('Ratings vs Favorite Counts')
plt.xlabel('Ratings')
plt.ylabel('Favorite Counts');
```

```python
plt.figure(figsize=(10,8))
plt.plot(ratings, retweet_counts, 'o', label='ratings vs retweets counts',color ='g')
plt.plot(ratings, intercept2 + slope2*ratings, 'r', label='fitted line')
plt.legend()
plt.title('Ratings vs Retweets Counts')
plt.xlabel('Ratings')
plt.ylabel('Retweets Counts')
```

#### Insights:

1, In terms of retweets number and favorites number, golden retriever, labrador retriever and pembroke being the top 3 most popular breeds

2, Average ratings have been increasing over the years, and it accelerated in 2017 showing that more and more high ratings dogs have appeared or users tended to give higher ratings recently 

3, There are positve correlations between ratings and retweet counts as well as ratings and favorite counts as shown in the graphs above. According to R squared, ratings and favorite counts has a stronger correlation than ratings and retweet counts.
